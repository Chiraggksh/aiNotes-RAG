```python
# ================= INSTALL LIBRARIES =================

!pip install -q \
youtube-transcript-api \
langchain \
langchain-community \
langchain-google-genai \
sentence-transformers \
faiss-cpu

# ================= IMPORTS =================

from youtube_transcript_api import (
    YouTubeTranscriptApi,
    TranscriptsDisabled
)

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI

from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import (
    RunnableParallel,
    RunnablePassthrough,
    RunnableLambda
)
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata

# ================= GOOGLE API KEY =================

GOOGLE_API_KEY = userdata.get("GOOGLE_API_KEY")

# ================= VIDEO ID =================
# Example:
# https://www.youtube.com/watch?v=Gfr50f6ZBvo

video_id = "EuBBz3bI-aA"

# ================= GET TRANSCRIPT =================

try:
    api = YouTubeTranscriptApi()

    transcript_data = api.fetch(video_id)

    transcript = " ".join(
        chunk.text
        for chunk in transcript_data
    )

    print("Transcript Loaded Successfully!")

except TranscriptsDisabled:
    print("No captions available.")
    transcript = ""

# ================= TEXT SPLITTING =================

splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=250
)

chunks = splitter.create_documents([transcript])

print(f"Chunks Created: {len(chunks)}")

# ================= HUGGINGFACE EMBEDDINGS =================
# FREE embeddings (No Gemini quota issue)

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# ================= VECTOR STORE =================

vector_store = FAISS.from_documents(
    chunks,
    embedding_model
)

print("Vector DB Created!")

# ================= RETRIEVER =================

retriever = vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 4}
)

# ================= GEMINI MODEL =================

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.2
)

# ================= PROMPT =================

prompt = PromptTemplate(
    template="""
You are a helpful assistant.

Answer ONLY using the transcript context.

If the information is not available
in the transcript, say:

"I don't know based on the video transcript."

Transcript Context:
{context}

Question:
{question}
""",
    input_variables=["context", "question"]
)

# ================= FORMAT DOCS =================

def format_docs(retrieved_docs):
    return "\n\n".join(
        doc.page_content
        for doc in retrieved_docs
    )

# ================= PARALLEL CHAIN =================

parallel_chain = RunnableParallel({
    "context": retriever | RunnableLambda(format_docs),
    "question": RunnablePassthrough()
})

# ================= OUTPUT PARSER =================

parser = StrOutputParser()

# ================= FINAL RAG CHAIN =================

main_chain = (
    parallel_chain
    | prompt
    | llm
    | parser
)

# ================= ASK QUESTIONS =================

question = "Can you summarize this video?"

answer = main_chain.invoke(question)

print("\n===== ANSWER =====\n")
print(answer)

# ================= CUSTOM QUESTIONS =================

question2 = "How does stat quest starts the video?"

answer2 = main_chain.invoke(question2)

print("\n===== SECOND QUESTION =====\n")
print(answer2)

question3 = "Where it is used and what is the conclusion?"

answer3 = main_chain.invoke(question3)

print("\n===== THIRD QUESTION =====\n")
print(answer3)
```
