```python
# Install LangChain
!pip install -q langchain langchain-text-splitters

# Import CharacterTextSplitter
from langchain_text_splitters import CharacterTextSplitter

# Sample text
text = """
Artificial Intelligence (AI) is one of the fastest-growing fields in technology.
It focuses on creating systems that can perform tasks requiring human intelligence,
such as decision-making, problem-solving, speech recognition, and image understanding.

Machine Learning (ML) is a branch of AI where machines learn patterns from data
instead of being explicitly programmed. For example, recommendation systems on
Netflix and YouTube use machine learning to suggest content based on user behavior.

Deep Learning is a specialized subset of machine learning that uses neural networks
with many hidden layers. It powers advanced technologies such as self-driving cars,
medical diagnosis systems, facial recognition, and virtual assistants like Siri and Alexa.

Natural Language Processing (NLP) is another important branch of AI.
It enables computers to understand, interpret, and generate human language.
Applications include chatbots, language translation, sentiment analysis, and text summarization.

Large Language Models (LLMs) such as GPT and Gemini are revolutionizing AI by
allowing systems to generate human-like responses, answer questions, write code,
summarize documents, and even assist in research and education.

LangChain is a framework that helps developers build applications powered by LLMs.
It provides tools for prompt engineering, document loading, text splitting,
vector databases, memory management, and retrieval-based question answering systems.

Text splitting is an important concept because LLMs have context limits.
Very large documents cannot be processed all at once, so they are divided into smaller chunks.
Chunk overlap helps preserve context between chunks by repeating a small portion
of the previous chunk in the next one.
"""

splitter = CharacterTextSplitter(
    separator="",       # split by every character
    chunk_size=100,
    chunk_overlap=10
)

chunks = splitter.split_text(text)

for i, chunk in enumerate(chunks):
    print(f"\n===== CHUNK {i+1} =====")
    print(repr(chunk))
```

```python
# Install package
!pip install -q langchain-text-splitters

# Import
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Long sample text
text = """
Artificial Intelligence (AI) is one of the fastest-growing fields in technology.
It focuses on creating systems that can perform tasks requiring human intelligence,
such as decision-making, problem-solving, speech recognition, and image understanding.

Machine Learning (ML) is a branch of AI where machines learn patterns from data.
Instead of being explicitly programmed, systems improve through experience and examples.
Recommendation systems on Netflix and YouTube are examples of machine learning applications.

Deep Learning is a subset of machine learning that uses neural networks with multiple layers.
It powers self-driving cars, facial recognition systems, medical diagnosis tools,
and virtual assistants like Siri and Alexa.

Natural Language Processing (NLP) helps computers understand human language.
Applications include chatbots, sentiment analysis, language translation,
and text summarization.

LangChain helps developers build applications powered by LLMs.
It provides tools for prompt templates, document loading,
text splitting, vector databases, and memory systems.
"""

# Create Recursive Text Splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,        # max chunk size
    chunk_overlap=50       # overlap between chunks
)

# Split text
chunks = splitter.split_text(text)

# Print chunks
for i, chunk in enumerate(chunks):
    print(f"\n{'='*20} CHUNK {i+1} {'='*20}")
    print(chunk)
    print("\nLength:", len(chunk))
```

```python
# Install package
!pip install -q langchain-text-splitters

# Import
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    Language
)

# ---------------- PYTHON CODE ----------------

python_code = """
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

class Calculator:
    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            return "Cannot divide by zero"
        return a / b

for i in range(5):
    print(i)
"""

# Python code splitter
python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=100,
    chunk_overlap=20
)

python_chunks = python_splitter.split_text(python_code)

print("="*20, "PYTHON SPLITS", "="*20)

for i, chunk in enumerate(python_chunks):
    print(f"\nCHUNK {i+1}")
    print(chunk)


# ---------------- MARKDOWN TEXT ----------------

markdown_text = """
# Artificial Intelligence

Artificial Intelligence (AI) is transforming industries.

## Machine Learning

Machine Learning allows systems to learn from data.

### Applications
- Recommendation Systems
- Chatbots
- Fraud Detection

## Deep Learning

Deep Learning uses neural networks to solve complex problems.

### Benefits
1. Image Recognition
2. Speech Recognition
3. NLP Applications
"""

# Markdown splitter
markdown_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.MARKDOWN,
    chunk_size=100,
    chunk_overlap=20
)

markdown_chunks = markdown_splitter.split_text(markdown_text)

print("\n")
print("="*20, "MARKDOWN SPLITS", "="*20)

for i, chunk in enumerate(markdown_chunks):
    print(f"\nCHUNK {i+1}")
    print(chunk)
```

```python
# Install libraries
!pip install -qU langchain langchain-community langchain-experimental langchain-google-genai

# Imports
from langchain_experimental.text_splitter import SemanticChunker
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from google.colab import userdata

# Get API key from Colab secrets
GOOGLE_API_KEY = userdata.get("GOOGLE_API_KEY")

# Working Gemini embedding model
embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001",
    google_api_key=GOOGLE_API_KEY
)

# Text with different semantic topics
text = """
Artificial Intelligence is transforming industries.
AI is used in robotics, healthcare, and automation.

Machine Learning is a subset of AI.
It helps computers learn patterns from data.

Cricket is one of the most popular sports in India.
IPL and the Cricket World Cup are major tournaments.

Football is loved worldwide.
Messi and Ronaldo are among the most famous players.

Java is one of the most popular programming languages.
It is widely used for backend development and DSA.
"""

# Create semantic splitter
splitter = SemanticChunker(
    embeddings=embeddings,
    breakpoint_threshold_type="percentile"
)

# Split semantically
docs = splitter.create_documents([text])

# Print chunks
for i, doc in enumerate(docs):
    print(f"\n{'='*20} CHUNK {i+1} {'='*20}")
    print(doc.page_content)
```
