```python
# Install required libraries
!pip install -q langchain langchain-community requests==2.32.4

# Import TextLoader
from langchain_community.document_loaders import TextLoader

# Create an example text file automatically
sample_text = """
Hello Chirag!

Welcome to LangChain TextLoader example.

LangChain helps us load documents such as:
- TXT files
- PDFs
- Web pages
- Word documents

This is a sample text file for testing in Google Colab.
"""

with open("sample.txt", "w", encoding="utf-8") as file:
    file.write(sample_text)

# Load the text document
loader = TextLoader("sample.txt", encoding="utf-8")
documents = loader.load()

# Print content
print("TEXT CONTENT:\n")
print(documents[0].page_content)

# Print metadata
print("\nMETADATA:\n")
print(documents[0].metadata)
```

```python
# Install required libraries
!pip install -q langchain langchain-community pypdf

# Import PyPDFLoader
from langchain_community.document_loaders import PyPDFLoader

# PDF path
pdf_path = "/content/2nd-year-__Scheme-and-Syllabus_AY-2022-23-onwards-of-B.Tech-AIML_AIDS_IIOT (1).pdf"

# Load PDF
loader = PyPDFLoader(pdf_path)
docs = loader.load()

# Total pages
print("Total Pages:", len(docs))

# First page content
print("\n===== FIRST PAGE CONTENT =====\n")
print(docs[0].page_content)

# Metadata
print("\n===== METADATA =====\n")
print(docs[0].metadata)
```

LOAD vs LAZY LOAD?
kaam same hota h but alg internally: load eager loading krta h sarii ek sathh
but lazy loader rather than list gives generator ek baar me sirf ek document load hota h and then uspee aap alg function lgatee ho

so load: eager loading, list document output, immediately into memory and best when small doc + want everything loaded upfront
but lazy loader mee : generator + used when large doc, stream processing eg chunking, embedding

from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    path='books',
    glob='*.pdf',
    loader_cls=PyPDFLoader
)

docs = loader.lazy_load() //if load use krta toh bht late ata phle prepare hota then return hotaaa

for document in docs:
    print(document.metadata) //here no time taking and aram se ajaegaaaa

```python
# Install required libraries
!pip install -q langchain langchain-community beautifulsoup4 lxml

# Import WebBaseLoader
from langchain_community.document_loaders import WebBaseLoader

# Website URL
url = "https://takeuforward.org/dsa/strivers-a2z-sheet-learn-dsa-a-to-z"

# Load webpage content
loader = WebBaseLoader(url)
docs = loader.load()

# Print webpage content
print("===== WEBSITE CONTENT =====\n")
print(docs[0].page_content[:3000])  # first 3000 characters

# Metadata
print("\n===== METADATA =====\n")
print(docs[0].metadata)
```
